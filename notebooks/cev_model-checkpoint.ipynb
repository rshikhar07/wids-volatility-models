```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import brentq

# Reproducibility
np.random.seed(42)
```

```python
def bs_call(S, K, r, sigma, T):
    """
    Black-Scholes price of a European Call option
    """
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    call_price = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price
```

```python
def simulate_cev(S0, r, sigma, gamma, T, n_paths, n_steps):
    """
    Simulate paths under the CEV model using Euler-Maruyama
    """
    dt = T / n_steps
    S = np.zeros((n_paths, n_steps + 1))
    S[:, 0] = S0

    for t in range(n_steps):
        Z = np.random.randn(n_paths)
        S[:, t + 1] = (
            S[:, t]
            + r * S[:, t] * dt
            + sigma * (np.maximum(S[:, t], 1e-8) ** gamma) * np.sqrt(dt) * Z
        )

    return S
```

```python
# Parameters
S0 = 100
r = 0.02
sigma = 0.2
T = 1
n_paths = 10      # small for visualization
n_steps = 252

# Simulate paths for gamma = 0.5
paths = simulate_cev(S0, r, sigma, gamma=0.5, T=T, n_paths=n_paths, n_steps=n_steps)

# Plot paths
plt.figure(figsize=(10, 5))
for i in range(n_paths):
    plt.plot(paths[i], alpha=0.7)

plt.title("CEV Model Paths (gamma = 0.5)")
plt.xlabel("Time Steps")
plt.ylabel("Stock Price")
plt.grid(True)
plt.show()
```

```python
# Monte Carlo pricing parameters
K = 100
n_paths_price = 50000   # as suggested by mentor
n_steps_price = 252

# Simulate paths
paths_price = simulate_cev(
    S0=S0,
    r=r,
    sigma=sigma,
    gamma=0.5,
    T=T,
    n_paths=n_paths_price,
    n_steps=n_steps_price
)

# Terminal prices
ST = paths_price[:, -1]

# Payoff and price
payoff = np.maximum(ST - K, 0)
price = np.exp(-r * T) * np.mean(payoff)

price
```

```python
def implied_vol_call(target_price, S, K, r, T):
    """
    Robust implied volatility solver with safety checks
    """
    def f(vol):
        return bs_call(S, K, r, vol, T) - target_price

    # Try wide bounds
    vol_low = 1e-4
    vol_high = 5.0

    # Check signs
    if f(vol_low) * f(vol_high) > 0:
        return np.nan  # implied vol not well-defined

    return brentq(f, vol_low, vol_high)
```

```python
# Mini volatility skew: 3 strikes

strikes = [80, 100, 120]
implied_vols = []
prices = []
```

```python
for K_i in strikes:
    # Simulate paths
    paths_i = simulate_cev(
        S0=S0,
        r=r,
        sigma=sigma,
        gamma=0.5,
        T=T,
        n_paths=n_paths_price,
        n_steps=n_steps_price
    )
    
    ST_i = paths_i[:, -1]
    payoff_i = np.maximum(ST_i - K_i, 0)
    price_i = np.exp(-r * T) * payoff_i.mean()
    
    # Implied volatility
    iv_i = implied_vol_call(price_i, S0, K_i, r, T)
    
    prices.append(price_i)
    implied_vols.append(iv_i)
    
    print(f"K={K_i}: Price={price_i:.4f}, Implied Vol={iv_i:.4f}")
```

```python
valid_strikes = []
valid_ivs = []

for k, iv in zip(strikes, implied_vols):
    if not np.isnan(iv):
        valid_strikes.append(k)
        valid_ivs.append(iv)
```

```python
plt.figure(figsize=(7, 5))
plt.plot(valid_strikes, valid_ivs, marker='o')
plt.xlabel("Strike")
plt.ylabel("Implied Volatility")
plt.title("Mini Volatility Skew (CEV, gamma = 0.5)")
plt.grid(True)
plt.show()
```
